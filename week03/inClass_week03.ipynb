```python
# what is the mean, max, min, std of square footage for each year aquired?
stats = buildings.groupby("Year Acquired")["Square Footage"].describe()
```

```python
stats
```

```python
type(stats)
```

```python
fig, ax = plt.subplots(figsize=(10,3))
ax.plot(stats.index,stats['max'], marker='.')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,3))
ax.plot(stats['max'], marker='.')
ax.set_xlabel('Year Acquired')
ax.set_ylabel('Max Square Footage')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,3))
ax.plot(stats['max'], marker='.')
ax.set_xlabel('Year Acquired')
ax.set_ylabel('Max Square Footage')
ax.set_yscale('log')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,3))
ax.plot(stats['max'], marker='.')
ax.plot(stats['min'], marker='.')
ax.set_xlabel('Year Acquired')
ax.set_ylabel('Max Square Footage')
ax.set_yscale('log')
plt.show()
```
